```python
# Strings	DNA sequence as text - Dictionaries	Store counts, Built-in functions	.count(), len(), Math:	GC content, Visualization

# Step 1 Create a fake DNA sequence
dna_sequence = "ATGCGATCGATCGTACGATCGCGATCGTAGCTAGCTAGCTGACT"

# Step 2: Count nucleotides
nucleotide_counts = {
    'A': dna_sequence.count('A'),
    'T': dna_sequence.count('T'),
    'C': dna_sequence.count('C'),
    'G': dna_sequence.count('G')
}

print("Nucleotide Counts:")
print(nucleotide_counts)

# Step 3: Calculate GC content
gc_content = (nucleotide_counts['G'] + nucleotide_counts['C']) / len(dna_sequence) * 100
print(f"GC Content: {gc_content:.2f}%")

# Step 4: Visualize
import matplotlib.pyplot as plt

plt.bar(nucleotide_counts.keys(), nucleotide_counts.values(), color=['orange', 'green', 'blue', 'red'])
plt.title("Nucleotide Composition in DNA Sequence")
plt.ylabel("Count")
plt.xlabel("Nucleotide")
plt.show()
```
