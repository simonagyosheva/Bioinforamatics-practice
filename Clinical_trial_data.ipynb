```python
# Create and explore dataset, visualise group comparison - data structure, dataframe, visualization
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Create mock patient data
data = {
    'Patient_ID': range(1, 11),
    'Age': [34, 45, 29, 54, 40, 36, 50, 60, 47, 38],
    'Treatment': ['A', 'B', 'A', 'B', 'A', 'A', 'B', 'B', 'A', 'B'],
    'Outcome': [1, 0, 1, 0, 1, 0, 0, 0, 1, 1]  # 1 = success, 0 = failure
}
df = pd.DataFrame(data)

# Step 3: Preview the data
df.head()

# Step 4: Summary statistics
print(df.describe())

# Step 5: Visualize outcome by treatment group
sns.countplot(x='Treatment', hue='Outcome', data=df)
plt.title("Treatment vs Outcome")
plt.show()
```
